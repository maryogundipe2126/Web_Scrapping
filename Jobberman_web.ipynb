```python
import csv
from datetime import datetime, timedelta
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
template = 'https://www.jobberman.com/jobs?sort=featured'
```

```python
def get_url(position):
    template = 'https://www.jobberman.com/jobs?sort=featured'
    url = template.format(position)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
```

```python
def transform(soup):
       
    divs = soup.find_all('div', class_="mx-5 md:mx-0 flex flex-wrap col-span-1 mb-5 bg-white rounded-lg border border-gray-300 hover:border-gray-400 focus-within:ring-2 focus-within:ring-offset-2 focus-within:ring-gray-500")
    for item in divs:
        title = item.find('a').text.strip()
        company = item.find('p', class_ ='text-sm text-link-500').text.strip()
        row = item.find_all('span', class_="mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide")
        if len(row) <2:
            continue
        location = row[0].text.strip()
        jobtype = row[1].text.strip()
        try:
            salary =item.find('span', class_="mr-1").text.strip()
        except:
            salary = 'Not given'
        function = item.find('p', class_="text-sm text-gray-500 text-loading-animate inline-block").text.strip().replace('\n', '').replace('Job Function:', '')
        date = item.find('p', class_ ="ml-auto text-sm font-normal text-gray-700 text-loading-animate").text
        today = datetime.today()
        date_str = re.search(r'\d+', date).group()
        extractdate = today - timedelta(days=int(date_str))
        posted_date = extractdate.strftime('%d %b %Y')
        summary = item.find('p', class_="text-sm font-normal text-gray-700 md:text-gray-500 md:pl-5").text.replace('\n', '')
        job = {
            'Title' : title,
            'Company' : company,
            'Location' : location,
            'Jobtype' : jobtype,
            'Salary': salary,
            'Job Function' : function,
             'Posted_on' :posted_date,
            'Job_requirement' : summary
            }
        jobpost.append(job)
    
    return 
jobpost = []
```

```python
c = get_url(0)
transform(c)
```
